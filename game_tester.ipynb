# A tester for the Game instance

```python
from brumaire.agent import RandomAgent
from brumaire.session import Game
from brumaire.renderer import Renderer
```

## Hyperparameters

```python
BOARD_NUM = 10
AGENTS = [
    RandomAgent(),
    RandomAgent(),
    RandomAgent(),
    RandomAgent(),
    RandomAgent()
]
```

## Start a game

Decide napoleons.

```python
game = Game(BOARD_NUM, AGENTS, log_enabled=True)
game.decide_napoleon()
```

```python
game.board.decl[0]
```

```python
game.board.roles[0]
```

```python
game.board.get_napoleon()[0]
```

Discard additional cards.

```python
game.discard_additional_cards()
```

Do tricks

```python
for idx in range(10):
    game.trick(idx)
```

```python
game.check_result()
```
